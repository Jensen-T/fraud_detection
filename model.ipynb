```python
data['Class'].value_counts()
```

The dataset is balanced, with an even number of fraudulent transactions (1) and valid transactions (0). This dataset contains anonymized features that have been processed via PCA, and no metadata exists for the original features. No null values exist. One thing to note is that in the real world, the likelihood for this proportion realistic vs fraudulent transactions to be observed.

```python
fraud = data[data['Class'] == 1]
normal = data[data['Class'] == 0]
```

We can have a basic look to identify if there are any transaction amount differences from fraudulent vs normal. All other features have been anonymized to protect cardholder identities.

```python
pd.concat([fraud.Amount.describe(), normal.Amount.describe()], axis=1)
```

When transaction amount is singled out, there is no noticable difference between fraud and normal transactions.
